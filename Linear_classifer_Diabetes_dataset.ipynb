```python
import pandas as pd
```

```python
diabetes = pd.read_csv("diabetes.csv")
```

```python
diabetes.head()
```

```python
diabetes.columns
```

```python
cols_to_norm = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']
```

```python
diabetes[cols_to_norm] = diabetes[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max()-x.min()) )
```

```python
diabetes.head()
```

```python
import tensorflow as tf
```

```python
diabetes.columns
```

```python
# these colums are for continous values 
num_preg = tf.feature_column.numeric_column('Pregnancies')
plasma_gluc = tf.feature_column.numeric_column('Glucose')
dias_press = tf.feature_column.numeric_column('BloodPressure')
tricep = tf.feature_column.numeric_column('SkinThickness')
insulin = tf.feature_column.numeric_column('Insulin')
bmi = tf.feature_column.numeric_column('BMI')
diabetes_pedigree = tf.feature_column.numeric_column('DiabetesPedigreeFunction')
age = tf.feature_column.numeric_column('Age')
```

```python
# now we need to deal with categorical values or non continous values 
# there are two ways to do this , 1 is using vocabulary list and 2nd is using hash bucket 
# so i know there are 4 possible groups i.e A, b ,c and d. so i will define my categ column, then i will pass a list
# of possible categories 
assigned_group = tf.feature_column.categorical_column_with_vocabulary_list("group", ["A", "B", "C", "D"])
```

```python
# if u r in the sitatution where u dont know all the grouping urself, for e.g name of all countries which is quite teddies
# task to write name of every country then you use hash_bucket and now instead of providing the list of possible categories, u 
# just define a hash bucket size which shows max amount of groups that u belive will be in that categoical column 
# tf automatically makes categgorical list 

# assigned_group = tf.feature_column.categorical_column_with_hash_bucket("Group", hash_bucket_size = 10)
```

```python
# now i m gonna discuss how to convert continous column into a categorcal column 
# i m gonna convert age into categorcail column  
```

```python
import matplotlib.pyplot as plt
```

```python
diabetes["Age"].hist(bins = 20)
```

```python
# overhere we r converting continous values into categoricl, here i will pass list of bucket i.e. 20 shows how many people are
# of the age 20 and how many are of 30 etc 
age_bucket = tf.feature_column.bucketized_column(age, boundaries= [20, 30, 40, 50, 60, 70, 80])
```

```python
# now i m gonna put all these things together.
# i m gonna make list of features column.
feat_cols = [num_preg, plasma_gluc, dias_press, tricep, insulin, bmi, diabetes_pedigree, assigned_group, age_bucket]
```

```python
# Train Test Split
```

```python
x_data = diabetes.drop("Outcome", axis = 1)
```

```python
x_data.head()
```

```python
labels = diabetes["Outcome"]
```

```python
#labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, labels, test_size= 0.3, random_state= 101 )
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x= x_train, y= y_train, batch_size = 10, num_epochs = 100, shuffle = True)
```

```python
x_test.shape
```

```python
model = tf.estimator.LinearClassifier(feature_columns= feat_cols, n_classes= 2)
```

```python
model.train(input_fn= input_func, steps= 100)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x= x_test, y= y_test, batch_size = 10, num_epochs = 1, shuffle = False
                                     )
```

```python
results = model.evaluate(eval_input_func)
```

```python
results
```

```python
# here x = "whaterver new data you have in"
# since we dont have any new data so we r going to make predictons on test data 
# and dont provide y value blc its prediction not evaluation
pred_input_func = tf.estimator.inputs.pandas_input_fn(x = x_test, batch_size= 10, num_epochs = 1, shuffle=False)
```

```python
predictions = model.predict(pred_input_func)
```

# Now Using Dense Neural Network Classifier

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units = [10, 10, 10], feature_columns = feat_cols,
                                      n_classes = 2)

# 10 neurons in each hidden layer
```

```python
# if u try it like before u did then u r going to get error
# in DNN we need to pass features column in the form of embedding columns
# blc the trouble is created by categorical column that we created 
# u need to pass ur categorical column into embedding column first
dnn_model.train(input_fn = input_func, steps= 1000)
```

```python
# in DNN we need to pass features column in the form of embedding columns
# blc the trouble is created by categorical column that we created 
# u need to pass ur categorical column into embedding column first
# dimension = 4 blc A, B,C and D 

embedded_group_col = tf.feature_column.embedding_column(assigned_group, dimension = 4)
```

```python
# once u done that u need to reset ur feature columns 
# so replacing assigned group with embedded_group_col

feat_cols = [num_preg, plasma_gluc, dias_press, tricep, insulin, bmi, diabetes_pedigree, embedded_group_col, age_bucket]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x_train, y_train, batch_size= 10, num_epochs= 1000, shuffle = True )
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units = [10, 10, 10], feature_columns= feat_cols, n_classes = 2)
```

```python
dnn_model.train(input_fn = input_func, steps = 1000)
```

```python
# in evaluation we pass num of epochs = 1 blc there is not point of checking accuracy of data again and again 
# in evaluation it is gonna compare the predicted value with the true label value. 
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test,y = y_test, batch_size= 10, num_epochs=1, shuffle = False)
```

```python
dnn_model.evaluate(eval_input_func)
```
